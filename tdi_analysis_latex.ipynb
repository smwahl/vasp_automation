```python
# make tdi dataframes that are not in tuple form for easier indexing
def unTuple(df):
    '''Function that takes a dataframe and converts any columns that have the format 
    of a tuple with length 2 into two columns, the first with the original column name
    and the second with a d + the original column name. Columns in this form contain 
    a value and its uncertainty
    '''
    newDf = pd.DataFrame(index=df.index)
    for col in df.columns:
        if isinstance(df[col][0],tuple) and len(df[col][0]) == 2:
            c0 =[ x[0] for x in df[col] ] 
            c1 =[ x[1] for x in df[col] ] 
            newDf[col] = c0
            newDf['d'+col] = c1
        else:
            newDf[col] = df[col]
    
    return newDf

#tdi_ut = unTuple(tdi)
```

```python
def estIsentrope(df,s_const):
    '''
    Estimate isentrope for a phase contained in datafram df, at a constant  entropy s_const
    '''
    g = pd.groupby(df,'P_target')
    #g.agg(lambda x: np.polyfit(x['temp'],x['S_dft'][0]))
    
    #s_const = .0016
    Tadiabat = []
    Padiabat = []
    for key, grp in g:
        grp1 = grp.sort('temp')
        t = grp1['temp']
        s = grp1['S_dft']
        
        # extend estimate below lowest temperature
        t0 = 0.
        s0 = s[0] - (s[1] - s[0])/(t[1] - t[0])*t[0]
        textend = np.array([ t0 ] + grp1['temp'].tolist())
        sextend = np.array([ s0] + grp1['S_dft'].tolist())
        
        Tadiabat.append(linInterp(sextend,textend,s_const))
        Padiabat.append(key)
    
    return np.array(Tadiabat), np.array(Padiabat)

Tad,Pad = estIsentrope(femgo_tdi2,.0025)
print Tad,Pad
```
