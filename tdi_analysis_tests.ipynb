```python
# make tdi dataframes that are not in tuple form for easier indexing
def unTuple(df):
    '''Function that takes a dataframe and converts any columns that have the format 
    of a tuple with length 2 into two columns, the first with the original column name
    and the second with a d + the original column name. Columns in this form contain 
    a value and its uncertainty
    '''
    newDf = pd.DataFrame(index=df.index)
    for col in df.columns:
        if isinstance(df[col][0],tuple) and len(df[col][0]) == 2:
            c0 =[ x[0] for x in df[col] ] 
            c1 =[ x[1] for x in df[col] ] 
            newDf[col] = c0
            newDf['d'+col] = c1
        else:
            newDf[col] = df[col]
    
    return newDf

#unTuple(tdi)
```

```python
# sum and store reaction energies
# Note: currently this requires there to be only one entry of a given formula at a given condition

rxn_list = []

for key in keys: # loop over conditions
#for key in [(50.,5000.)]:
    cond = tdi.loc[conds[key]] 
    
    for r in reactions:
        # stores energies for the reaction
        energies_list = []
        dft_ids = []
        systems = []
        mult_list = []
        
        for idx,row in cond.iterrows():
            sys = row['system']
            
            # include only sytems in the reaction
            if sys in r.keys():
                mult = mults[sys] * r[sys]
                mult_list.append(mult)
                arr = np.array(row[['F_dft','P_targetV','U_dft','G_dft','minusTS']].tolist())
                energies = arr * mult
                energies[:,1] = np.abs(energies[:,1]) # all errors positive
                energies_list.append(energies)
                dft_ids.append(row['dft_id'])
                systems.append(sys)
    
        # check that all systems in reaction are accounted for
        if(len(energies_list) == len(r.keys())): 
            # Sum the energies fro the given reaction
            rxn_energies = np.zeros([5,2])    
            for arr in energies_list:
                rxn_energies += arr

            rxn_energies_eV = rxn_energies*Ha_to_eV
            #dft_ids = cond['dft_id'].tolist()

            # categorize reaction
            #print dft_ids,systems
            rxn_type,rxn_label = catRxn(dft_ids,systems)            
            rxn_list.append( [key, rxn_type,rxn_label, key[0], key[1], dft_ids] + [ rxn_energies[i,:] for i in range(5) ] +  \
                [ rxn_energies_eV[i,:] for i in range(5) ] )
        else: 
            #rxn_list.append( [key, key[0], key[1], dft_ids] + [ None for i in range(10) ] )
            pass # don't add incomplete reactions
        
rxn = pd.DataFrame(rxn_list,columns=['cond','rxn_type','rxn_label','P_target','temp','dft_ids','F_dft','P_targetV','U_dft','G_dft','minusTS', \
                             'F_dft_eV','P_targetV_eV','U_dft_eV','G_dft_eV','minusTS_eV'] )
#rxn.set_index('cond',inplace=True,drop=False)
```

```python
def estIsentrope(df,s_const):
    '''
    Estimate isentrope for a phase contained in datafram df, at a constant  entropy s_const
    '''
    g = pd.groupby(df,'P_target')
    #g.agg(lambda x: np.polyfit(x['temp'],x['S_dft'][0]))
    
    #s_const = .0016
    Tadiabat = []
    Padiabat = []
    for key, grp in g:
        grp1 = grp.sort('temp')
        t = grp1['temp']
        s = grp1['S_dft']
        
        # extend estimate below lowest temperature
        t0 = 0.
        s0 = s[0] - (s[1] - s[0])/(t[1] - t[0])*t[0]
        textend = np.array([ t0 ] + grp1['temp'].tolist())
        sextend = np.array([ s0] + grp1['S_dft'].tolist())
        
        Tadiabat.append(linInterp(sextend,textend,s_const))
        Padiabat.append(key)
    
    return np.array(Tadiabat), np.array(Padiabat)

Tad,Pad = estIsentrope(femgo_tdi2,.0025)
print Tad,Pad
```

```python
# predict new volumes based on coeff. of thermal expansion

# Group tdi results by 
P_sys = tdi.groupby(['P_target','system'])

grps = P_sys.groups   

def newV(T,To,Vo,alpha):
    V = Vo + Vo * ( T - To) * alpha
    a = V**(1/3.)
    return V,a
alpha_data = {}

# store alpha and relavent data for finding new volumes
for key in grps.keys():
    runs = grps[key]
    T1 = tdi.loc[runs[0]]['temp'] 
    T2 = tdi.loc[runs[1]]['temp']
    V1 = tdi.loc[runs[0]]['volume']
    V2 = tdi.loc[runs[1]]['volume']
    alpha = ( V2 - V1 ) / float(V1) / float( T2 - T1 )
    alpha_data.update({key:[T1,V1,alpha]})
    
    print 'Alpha for {} : {}'.format(key,alpha)
    print '    V1 = {} , T1 = {}'.format(V1,T1)
```

```python
# find new volumes at new temperatures based on existing runs
for x,t in zip([(100.0, 'Fe32'),(100.0, 'O24Fe24Mg24'),(400.0, 'Fe32'),(400.0, 'O24Fe24Mg24')],
               [7000.,7000.,8000.,8000.]):
    a = alpha_data[x]
    print x, newV(t,a[0],a[1],a[2])
```
