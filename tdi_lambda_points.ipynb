```python
# make tdi dataframes that are not in tuple form for easier indexing
def unTuple(df):
    '''Function that takes a dataframe and converts any columns that have the format 
    of a tuple with length 2 into two columns, the first with the original column name
    and the second with a d + the original column name. Columns in this form contain 
    a value and its uncertainty
    '''
    newDf = pd.DataFrame(index=df.index)
    for col in df.columns:
        if isinstance(df[col][0],tuple) and len(df[col][0]) == 2:
            c0 =[ x[0] for x in df[col] ] 
            c1 =[ x[1] for x in df[col] ] 
            newDf[col] = c0
            newDf['d'+col] = c1
        else:
            newDf[col] = df[col]
    
    return newDf

#unTuple(tdi)
```

```python
# factors to multiply each system by
#mults_mgo = {'Fe32':-1/32.,'Mg32O32':-1/32.,'O24Fe24Mg24':1/24.}
#mults_sio = {'Fe32':-1/32.,'Si32O64':-1/32.,'Fe24Si24O48':1/24.}

# multiples to reduce to per formula unit (all positive)
mults = { 'Fe32':1/32.,'Mg32O32':1/32.,'O24Fe24Mg24':1/24.,'Si32O64':1/32.,'Fe24Si24O48':1/24.}

keys = conds.keys()

# reactions to identify (perhaps I should have a way of indexing by formula unit)
reactions = [{ 'Fe32':-1.,'Mg32O32':-1.,'O24Fe24Mg24':1.},{ 'Fe32':-1.,'Si32O64':-1.,'Fe24Si24O48':1.} ]

# to categorize reaction based on prefixes list of dft_ids
def catRxn(dft_ids):
    
    # change these to cateforize new reaction types
    rxn_sets = [{'lMgO','lFe','lFeMgO'},{'MgO','lFe','lFeMgO'},{'lSiO','lFeSiO','lFe'}]
    rxn_names = ['MgO_liq','MgO_sol','SiO2_liq']
    
    # remove the final 3 (numeric) characters to obtain the prefixes
    prefixes = set( [ x[:-3] for x in dft_ids ] )
    
    name = None
    for n,r in zip(rxn_names,rxn_sets):
        if set(prefixes) == set(r):
            name = n
            
    return name

#catRxn(['lMgO321','lFe123','lFeMgO444'])
```

```python
# sum and store reaction energies
# Note: currently this requires there to be only one entry of a given formula at a given condition

rxn_list = []

for key in keys: # loop over conditions
#for key in [(50.,5000.)]:
    cond = tdi.loc[conds[key]] 
    
    for r in reactions:
        # stores energies for the reaction
        energies_list = []
        dft_ids = []
        mult_list = []
        
        for idx,row in cond.iterrows():
            sys = row['system']
            
            # include only sytems in the reaction
            if sys in r.keys():
                mult = mults[sys] * r[sys]
                mult_list.append(mult)
                arr = np.array(row[['F_dft','P_targetV','U_dft','G_dft','minusTS']].tolist())
                energies = arr * mult
                energies[:,1] = np.abs(energies[:,1]) # all errors positive
                energies_list.append(energies)
                dft_ids.append(row['dft_id'])
    
        #print dft_ids
        #print mult_list
        #print energies_list
        # Sum the energies fro the given reaction
        rxn_energies = np.zeros([5,2])    
        for arr in energies_list:
            rxn_energies += arr

        rxn_energies_eV = rxn_energies*Ha_to_eV
        #dft_ids = cond['dft_id'].tolist()
        
        # categorize reaction
        rxn_type = catRxn(dft_ids)
    
        # check that all systems in reaction are accounted for
        if(len(energies_list) == len(r.keys())): 
           rxn_list.append( [key, rxn_type, row.P_target,row.temp, dft_ids] + [ rxn_energies[i,:] for i in range(5) ] +  \
                [ rxn_energies_eV[i,:] for i in range(5) ] )
        else: 
            #rxn_list.append( [key, key[0], key[1], dft_ids] + [ None for i in range(10) ] )
            pass # don't add incomplete reactions
        
rxn = pd.DataFrame(rxn_list,columns=['num_lambda','rxn_type','P_target','temp','dft_ids','F_dft','P_targetV','U_dft','G_dft','minusTS', \
                             'F_dft_eV','P_targetV_eV','U_dft_eV','G_dft_eV','minusTS_eV'] )
#rxn.set_index('cond',inplace=True,drop=False)
```
